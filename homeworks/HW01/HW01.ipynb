```python
K = np.ones((2,2,3,3)) #4 мерный массив
print("K "*20)
print(K)
print("Shape: ", K.shape)
print("Dimensions: ", K.ndim)
```

```python
print("k "*20)
arr = np.arange(1,37,1)
k = arr.reshape((2,2,3,3))
print(k)
```

```python
New_K = K+k
print("K+k "*10)
print(New_K)
```

```python
New_K2 = (K+K)*k
print("(K+K)*k "*10)
print(New_K2)
print(New_K2[1,1,2,2]) # должно быть значение 72 ((1+1)*36)
print(New_K2[1,0,0,0]) # должно быть значение 38 ((1+1)*19)
```
